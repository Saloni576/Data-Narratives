# Q2] What is the average number of break points created by players in a tennis match?

```python
data = pd.read_csv("/content/AusOpen-women-2013.csv")
data["BPC.1"], data["BPW.1"] = data["BPW.1"], data["BPC.1"]
data["BPC.2"], data["BPW.2"] = data["BPW.2"], data["BPC.2"]
data["NPA.1"], data["NPW.1"] = data["NPW.1"], data["NPA.1"]
data["NPA.2"], data["NPW.2"] = data["NPW.2"], data["NPA.2"]
data.head()
```

```python
total_bpc = data["BPC.1"] + data["BPC.2"]
avg_bpc = total_bpc.mean()
print("Average number of break points created per match:", avg_bpc)

plt.hist(total_bpc, bins=20, color='green')
plt.axvline(avg_bpc, color='red', linestyle='dashed', linewidth=2)
plt.xlabel("Break points created")
plt.ylabel("Number of matches")
plt.title("Distribution of break points created by players")
plt.show()
```

# Q3] Does a higher number of aces won by a player in a tennis match correspond to a higher probability of winning the match?

```python
data = pd.read_csv("/content/FrenchOpen-men-2013.csv")
data.head()
```

```python
winner_aces_df = data.loc[data['Result'] == 1, ['ACE.1']]
loser_aces_df = data.loc[data['Result'] == 0, ['ACE.2']]

avg_aces_winner = winner_aces_df.mean()[0]
avg_aces_loser = loser_aces_df.mean()[0]

plt.hist(winner_aces_df['ACE.1'], bins=20, alpha=0.5, label='Winner')
plt.hist(loser_aces_df['ACE.2'], bins=20, alpha=0.5, label='Loser')
plt.axvline(x=avg_aces_winner, color='blue', label='Avg. Aces Won by Winner')
plt.axvline(x=avg_aces_loser, color='orange', label='Avg. Aces Won by Loser')
plt.xlabel('Aces Won')
plt.ylabel('Frequency')
plt.title('Distribution of Aces Won by Winner and Loser')
plt.legend()
plt.show()
```

# Q4] What is the impact of winning or losing Set 1 on the outcome of a tennis match for a player?

```python
data = pd.read_csv("/content/FrenchOpen-women-2013.csv")
data.head()
```

```python
set1_df = data[['ST1.1', 'ST1.2', 'Result']].copy()
set1_df.dropna(inplace=True)
set1_df['Set1_winner'] = set1_df.apply(lambda x: x['ST1.1'] if x['Result'] == 1
                                       else x['ST1.2'], axis=1)
set1_df['Set1_loser'] = set1_df.apply(lambda x: x['ST1.2'] if x['Result'] == 1
                                      else x['ST1.1'], axis=1)
set1_df['Set1_winner_win'] = set1_df.apply(lambda x: 1
                                           if x['Set1_winner'] > x['Set1_loser']
                                           else 0, axis=1)
set1_win_percentage = set1_df.groupby('Set1_winner_win').size().div(len(set1_df)).mul(100).round(2)
labels = ['Lost Set 1', 'Won Set 1']
values = set1_win_percentage.values

plt.bar(labels, values, color="#DC143C")
plt.xlabel('Set 1 Result')
plt.ylabel('Percentage of Matches Won')
plt.title('Impact of Winning/Losing Set 1 on Match Outcome')
plt.show()
```

# Q5] How does the number of double faults committed by the loser of a tennis match compare to the average number of double faults in a match?

```python
data = pd.read_csv("/content/USOpen-men-2013.csv")
data.head()
```

```python
loser_df = data[['DBF.2']][data['Result'] == 0]

avg_df_match = data['DBF.1'].mean()
avg_df_loser = loser_df.mean()
colors = ['#FFC300', '#C70039']

plt.hist([data['DBF.1'].dropna(), loser_df['DBF.2'].dropna()], bins=20,
         alpha=0.8, label=['Overall', 'Loser'], color=colors)
plt.axvline(x=avg_df_match, color='#EEAD0E', label='Avg. Double Faults in Match')
plt.axvline(x=avg_df_loser[0], color='#FF3030', label='Avg. Double Faults by Loser')
plt.xlabel('Double Faults')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

# Q6] Is there a correlation between the number of net points attempted and the number of net points won by a player in a tennis match?

```python
data = pd.read_csv('/content/USOpen-women-2013.csv')
data.head()
```

```python
winner_net_points = data[['NPW.1', 'NPA.1']]

corr = winner_net_points.corr()

plt.scatter(winner_net_points['NPA.1'], winner_net_points['NPW.1'], color='purple')
plt.plot(winner_net_points['NPA.1'],
         winner_net_points['NPA.1']*corr.iloc[0,1] - corr.iloc[0,0], color='orange')
plt.xlabel('Net Points Attempted')
plt.ylabel('Net Points Won')
plt.title('Relationship between Net Points Attempted and Won by Winner')
plt.show()
```
